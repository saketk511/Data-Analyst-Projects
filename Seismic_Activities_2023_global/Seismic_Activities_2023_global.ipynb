# Global Seismic Activities 2023 Analysis

The dataset "Seismic Activities 2023 global" provides valuable information about earthquakes and other seismic activity worldwide in the year 2023. This dataset includes various parameters such as time, location (latitude and longitude), depth, magnitude, magnitude type, and more. In this exploration, we will delve into the dataset to gain insights into the seismic activity observed throughout the year.

```python
# Import the required libraries
import pandas as pd
import numpy as np
```

```python
# Read the data
data = pd.read_csv('Seismic_Activities_2023_global.csv')
data.head() # Helpful for a quick visual inspection of the dataset
```

```python
data.tail() # Sometimes there is aggregate row in end, we can check that by using this
```

```python
# Check the shape of data
data.shape # 26642 rows and 22 columns
```

```python
# Basic info about raw dataset
data.info() 
# Here we check, whether the data types assigned to respective columns are correct
# Which columns have null values
# Memory usuage
```

```python
# To get summarized analysis of numerical column
data.describe()
```

```python
# View the unique values
data['magType'].unique()
```

```python
data['type'].unique()
```

```python
data['id'].nunique() # To get the total number of unique 'id' values
```

```python
# Dealing with null values
data.isna().any() 
# Here there are frew columns which has null values i.e. nst, gap, dmin, place, horizontalError, magError and magNst.
```

```python
data.isna().sum()
```

```python
# Creating copy of data
data_temp = data.copy()
```

```python
# Filling Null values

data_temp['dmin'] = data_temp['dmin'].fillna(data_temp['dmin'].mean())
data_temp['place'] = data_temp['place'].replace(np.nan, 'Unknown')
data_temp['horizontalError'] = data_temp['horizontalError'].fillna(data_temp['horizontalError'].median())
data_temp['magError'] = data_temp['magError'].fillna(data_temp['magError'].median())
data_temp['magNst'] = data_temp['magNst'].fillna(33)
```

```python
data_temp.isna().sum()
```

```python
# Checking the distribution of data
kurtosis_value = data_temp.kurtosis() # Check kurtosis
print(kurtosis_value)
```

```python
skewness_value = data_temp.skew() # Check skew
print(skewness_value)
```

```python
# Removing outliers
# Finding out IQR for different variables.
Q1 = data_temp.quantile(0.25)
Q3 = data_temp.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Removing outliers
data_cleaned = data_temp[~((data_temp < (Q1 - 10 * IQR)) |(data_temp > (Q3 + 10 * IQR))).any(axis=1)] 
# Normally 1.5*IQR is taken but here I am taking 10*IQR  to accomodate some geniune values
```

```python
# Displaying shape before and after removing outliers
print("Shape with outliers: ", data_temp.shape)
print("Shape without outliers: ", data_cleaned.shape) # We have removed 394 outliers
```

```python
# Converting the cleaned data to csv
data_cleaned.to_csv("Earthquake 2023 cleaned data")
```
